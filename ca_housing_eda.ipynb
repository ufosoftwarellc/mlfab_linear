```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read in the cleaned data

```python
data_dir = 'data'
df = pd.read_csv(data_dir+'/calhousing_clean.csv')
df
```

## Dataframe info to look for missing values and the datatypes of the data
1. There are no missing values
2. All the data is numeric

```python
df.info()
```

## Dataframe describe to view the distribution of the data

```python
df.describe()
```

## Pair Plot to give a view of the relationship between the features
1. There is a clear linear increasing relation ship between MedInc and Price

```python
sns.set(rc={"figure.figsize":(1000, 1000)});
plt.style.use("dark_background");
sns.pairplot(df);
```

## Correlation Heatmap to view how correlated the features are
1. There is a strong positive correlation between MedInc and Price

```python
fig = px.imshow(df.corr(), text_auto=True,  template="plotly_dark")
fig.update_layout(
    title='Correlation Heatmap',
    title_x=0.5,
    width=1200,
    height=1200,
)

# use fig.show for interactive plot, use fig.show('png') so the plot renders in github
#fig.show()
fig.show('png')
```

## Histogram of the Price (the target)

```python
fig = px.histogram(df, x=df.Price, template="plotly_dark", histnorm='probability')
# use fig.show for interactive plot, use fig.show('png') so the plot renders in github
#fig.show()
fig.show('png')
```
